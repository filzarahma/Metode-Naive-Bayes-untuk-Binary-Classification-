# **Membaca Data Latih Uji**

```python
#import library
import pandas as pd
```

```python
#membaca data latih/uji
excel = pd.ExcelFile('traintest.xlsx')
tabel_train = pd.read_excel(excel, 'train')
print("----Data Train----")
print(tabel_train)
```

```python
#mengubah format tabel train ke bentuk dictionary
data_train = tabel_train.to_dict('records')
print(data_train)
```

# **Data Training**

```python
#statistika
#kamus banyaknya data
tabY = {'0': 0, '1': 0}
tabX1 = {}
tabX2 = {}
tabX3 = {}

#tambah banyak data
def countX(x, y, tab):
  if (tab.get(x) == None):
    tab[x] = {'0': 0, '1': 0}
  tab[x][y] += 1
  return tab

#menghitung banyaknya data per seluruh kategori
for i in range(0, len(data_train)):
  y = data_train[i]['y']
  x1 = data_train[i]['x1']
  x2 = data_train[i]['x2']
  x3 = data_train[i]['x3']
  
  if y == 0:
    tabY['0'] += 1
  else:
    tabY['1'] += 1

  tabX1 = countX(x1, str(y), tabX1)
  tabX2 = countX(x2, str(y), tabX2)
  tabX3 = countX(x3, str(y), tabX3)

#mengurutkan dictionary berdasarkan key
tabX1 = dict(sorted(tabX1.items()))
tabX2 = dict(sorted(tabX2.items()))
tabX3 = dict(sorted(tabX3.items()))

def statDataset(x, data):
  print()
  print(x," :")
  print("+-------+---------------+")
  print("|\t|\t y = \t|")
  print("| ", x,"\t+-------+-------+")
  print("| ","\t| 0 \t| 1\t|")
  print("+-------+---------------+")
  key = list(data.keys())
  for i in range(0, len(key)):
      print("|", key[i],"\t|", data[key[i]]['0'] ,"\t|", data[key[i]]['1'],"\t|")
  print("+-------+---------------+")

print("STATISTIKA DATASET")
print()
print("y  :")
print("-----------------")
print("| y", "\t| Jml", "\t|")
print("-----------------")
print("| 0", "\t|", tabY['0'], "\t|")
print("| 1", "\t|", tabY['1'], "\t|")
print("-----------------")

statDataset("x1", tabX1)
statDataset("x2", tabX2)
statDataset("x3", tabX3)
```

```python
#laplacian correction
def laplacianCorrectionY(tab):
  tab['0'] += 1
  tab['1'] += 1
  return tab

def laplacianCorrectionX(tab):
  k = list(tab.keys())
  for i in range(0, len(k)):
    tab[k[i]]['0'] += 1
    tab[k[i]]['1'] += 1
  return tab

tabY = laplacianCorrectionY(tabY)
tabX1 = laplacianCorrectionX(tabX1)
tabX2 = laplacianCorrectionX(tabX2)
tabX3 = laplacianCorrectionX(tabX3)

print("STATISTIKA DATASET SETELAH MENGGUNAKAN LAPLACIAN CORRECTION")
print()
print("y  :")
print("-----------------")
print("| y", "\t| Jml", "\t|")
print("-----------------")
print("| 0", "\t|", tabY['0'], "\t|")
print("| 1", "\t|", tabY['1'], "\t|")
print("-----------------")

statDataset("x1", tabX1)
statDataset("x2", tabX2)
statDataset("x3", tabX3)
```

# **Hasil Training Model**

```python
#menyimpan model hasil training
Py = {'0': 0, '1': 0}
Px1y = {}
Px2y = {}
Px3y = {}

#menghitung peluang
def peluang(nA, nS):
  return nA / nS

#menghitung banyaknya angka yang memiliki y = 0 dan  y = 1
def n(tab, key):
  total0 = 0
  total1 = 0
  for j in range(0, len(key)):
    total0 += tab[key[j]]['0']
    total1 += tab[key[j]]['1']
  return total0, total1

#peluang prior untuk kelas y
Py['0'] = peluang(tabY['0'], (tabY['0'] + tabY['1'])) #P(y=0)
Py['1'] = peluang(tabY['1'], (tabY['0'] + tabY['1'])) #P(y=1)

#peluang likelihood (X1 | y)
Px1y = tabX1
keyX1 = list(Px1y.keys()) 
a, b = n(tabX1, keyX1)
for i in range(0, len(keyX1)):
  Px1y[keyX1[i]]['0'] = peluang(tabX1[keyX1[i]]['0'], a) #P(X1| y=0)
  Px1y[keyX1[i]]['1'] = peluang(tabX1[keyX1[i]]['1'], b) #P(X1| y=1)

#peluang likelihood (X2 | y)
Px2y = tabX2
keyX2 = list(Px2y.keys())
a, b = n(tabX2, keyX2)
for i in range(0, len(keyX2)):
  Px2y[keyX2[i]]['0'] = peluang(tabX2[keyX2[i]]['0'], a) #P(X2| y=0)
  Px2y[keyX2[i]]['1'] = peluang(tabX2[keyX2[i]]['1'], b) #P(X2| y=1)

#peluang likelihood (X3 | y)
Px3y = tabX3
keyX3 = list(Px3y.keys())
a, b = n(tabX3, keyX3)
for i in range(0, len(keyX3)):
  Px3y[keyX3[i]]['0'] = peluang(tabX3[keyX3[i]]['0'], a) #P(X3| y=0)
  Px3y[keyX3[i]]['1'] = peluang(tabX3[keyX3[i]]['1'], b) #P(X3| y=1)

#menampilkan hasil training model
print("Hasil training model")

print()
print("Tabel y")
print("---------------------------------")
print("| y", "\t| Probabilitas", "\t\t|")
print("---------------------------------")
print("| 0", "\t|", Py['0'], "\t|")
print("| 1", "\t|", Py['1'], "\t|")
print("---------------------------------")

def printTable(x, Pxy, key):
  print()
  print("Tabel", x)
  print("+-------+---------------------------------------------------------------+")
  print("|\t|\t\t\t Probabilitas y = \t\t\t|")
  print("| ", x,"\t+-------------------------------+-------------------------------+")
  print("| ","\t|\t\t 0 \t\t|\t\t 1\t\t|")
  print("+-------+---------------------------------------------------------------+")
  for i in range(0, len(key)):
    print('{:<3}{:<5}{:<3}{:<29}{:<3}{:<29}{:<3}'.format("|", key[i],"|", Pxy[key[i]]['0'], "|" , Pxy[key[i]]['1'],"|"))
  print("+-------+---------------------------------------------------------------+")

printTable("X1", Px1y, keyX1)
printTable("X2", Px2y, keyX2)
printTable("X3", Px3y, keyX3)
```

# **Testing Model**

```python
#membaca data test
print("----Data Test----")
tabel_test = pd.read_excel(excel, 'test')
print(tabel_test)
```

```python
#mengubah format tabel test ke bentuk dictionary
data_test = tabel_test.to_dict('records')
print(data_test)
```

```python
#testing model

def NaiveBayes(X, Px1_y, Px2_y, Px3_y, Py):
  X_y0 = Px1_y[X['x1']]['0'] * Px2_y[X['x2']]['0'] * Px3_y[X['x3']]['0'] #P(X|y=0)
  X_y1 = Px1_y[X['x1']]['1'] * Px2_y[X['x2']]['1'] * Px3_y[X['x3']]['1'] #P(X|y=1)

  if X_y0 * Py['0'] > X_y1 * Py['1']:
    hasil = 0
  else :
    hasil = 1
  return hasil

#menghitung hasil test sekaligus menyimpan di data_test

for j in range(0, len(data_test)):
  data_test[j]['y'] = NaiveBayes(data_test[j], Px1y, Px2y, Px3y, Py)

print(data_test)
```

# **Final**

```python
#menyimpan output ke file
pd.DataFrame(data_test).to_excel('hasil.xlsx', sheet_name = 'test')
```

# **Akurasi**

```python
#menghitung akurasi algoritma

def akurasi(real, Px1y, Px2y, Px3y, Py):
  n = 0
  for i in range(len(real)):
    if NaiveBayes(real[i], Px1y, Px2y, Px3y, Py) == real[i]['y']:
      n += 1
  
  print("banyak data yang akurat adalah", n, "dari", len(real))
  print("akurasi data =", n, "/", len(real), "=", n/len(real)*100, "%")

akurasi(data_train, Px1y, Px2y, Px3y, Py)
```
